```python
from datafog import DataFog

sotu_url = 'https://gist.githubusercontent.com/sidmohan0/1aa3ec38b4e6594d3c34b113f2e0962d/raw/42e57146197be0f85a5901cd1dcdd9ad15b31bab/sotu_2023.txt'

# Create a DataFog instance with PII detection enabled and pointing to the SOTU URL
datafog = DataFog(pii_detection=True, source_url=sotu_url)
# Process the text and get the response
response = datafog.process_text()
# print("Extracted Text:", response.text)
print("Entities Detected:", response.entities)
```
